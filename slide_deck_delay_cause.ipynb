```python
# import libriaries #
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Read the dataset #
df = pd.read_csv('Airline_delay_causes17_20.csv')
```

```python
base_color = sns.color_palette()[0] # Indentify the first color from the tuple list of colors
# The below function will help throughtout this project , as we have to mention X - lable , y-lable , 
# title each and every time we plot any graph for clear understanding .
def lables_name(x , y = None):
    plt.xlabel(x)
    plt.ylabel(y)
    if(y == None):
        plt.title(f' {x} Distribution');
    else:
        plt.title(f' {x} and {y} Distribution')
```

```python
# Check which data will be picked for analysis and which one will be dropped
plt.figure(figsize = [15, 15])
sns.heatmap(df.corr(), annot = True, fmt = '.3f', center = 0)
lables_name('Correlation')
plt.show()
```

```python
### From the first look we can identify the variables will be taken into consideration.
### Drop the unusful column named 'Unnamed 21'
df_cause=df[['year','carrier','carrier_name',' month','airport','airport_name',' arr_delay',' carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']]
df_cause= df_cause.dropna(axis='rows')
```

```python
# Extracting the features from the original data and create a new dataframe for further implementation .
df_cause_only=df_cause[[' carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']]
```

### Everyone one of us always has his plans for holidays and traveling comes on the top of entertaining options so let me ask you
### How many times your plans for Christmas or a wedding party destroyed because of Airlines delay? let me tell some numbers based on a data from 2017 to 2020

### What if I told you that I can tell you the main cause of the delay so you can avoid it in the future?

```python
plt.figure(figsize = [8,8])
delay=df_cause_only.sum()
sns.barplot(delay, delay.index.values, color = base_color);
lables_name('Delay Variables');
```

### So my friend as you see the main cause of delay is aircraft delay and then carrier delay 
### so in the future be carefull when you choose the carrier 

## What about the most carrier in delay occurance? 

```python
plt.figure(figsize = [8,8])
carrier=df_cause['carrier_name'].value_counts()
sns.barplot(carrier, carrier.index.values, color = base_color);
lables_name('Carrier Variables');
```

### The Majority of delay comes from SkyWest Airlines With a percentage 15%.
### So of course in your future traveling plans use Virgin America it is more safe

## What about Destinations which one has the main share?

```python
plt.figure(figsize = [8,8])
destination=df_cause['airport_name'].value_counts()[:10]
sns.barplot(destination, destination.index.values, color = base_color);
lables_name('Carrier Variables');
```

### Actually I can't tell you that there is a certain airport cause delay they are almost the same

### but I have a good news for the delay causes records is decreasing as you can see below

```python
sns.regplot(data = df_cause , x= 'year' , y = 'late_aircraft_delay');
lables_name('Year' , 'late_aircraft_delay')
```

> The dealy causes is declining it's tendes to minimum in 2020

## I can also tell about the best time/worst time to travel, Imagine that I am your free travel agent now.

```python
sns.barplot(data = df_cause, x = ' month', y = 'weather_delay',color = base_color);
plt.xticks(rotation = 15);
lables_name('Month' , 'Weather delay')
```

```python
sns.barplot(data = df_cause, x = ' month', y = ' carrier_delay',color = base_color);
plt.xticks(rotation = 15);
lables_name('Month' , 'Carrier delay')
```

#### The worest time to travel is from June to August.
